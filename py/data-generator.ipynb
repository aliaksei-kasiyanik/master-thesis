```python
def time_to_min(str_time):
    t = strptime(str_time, '%H:%M')
    return t.tm_hour * 60 + t.tm_min
```

```python
def min_to_time(minutes):
    return "" + str(minutes // 60) + ":" + str(minutes % 60)
```
